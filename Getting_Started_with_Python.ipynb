# Data Types in Python

1. Numeric data type :- Int , Float

2. String :- sequence of character enclosed with single quote', double quote " or triple quote '''.

 **Creating Strings**  

 ● Single Quotes: 'Hello'

 ● Double Quotes: "Python"

 ● Triple Quotes: '''Multi-line String'''

3. Lists (list) :- A list is an ordered collection of items, which can be of different types. Lists are mutable, meaning they can be changed after creation.

4. Tuples (tuple) :- A tuple is a immutable collection of items meaning they cannot be changed after creation.

 **Why Use Tuples Instead of Lists?**

● Tuples are faster than lists.

● Useful for data that should not be changed (e.g., coordinates, fixed
configurations).

5. Set - is a unordered collection of unique data elements

 Set Operations

● Union (|) – Combines two sets.

● Intersection (&) – Returns common elements.

● Difference (-) – Finds elements in one set but not another.

6. Dictionaries (dict) :- unodered collection of key value pair data item

Built-in Functions for Data Types

 Type Conversion Functions

● int() – Converts to an integer.

● float() – Converts to a float.

● str() – Converts to a string.

● list() – Converts to a list.

● tuple() – Converts to a tuple.

● set() – Converts to a set.

● dict() – Converts to a dictionary.

```python
# Numeric Data Types
x = 20
y = 15.6
c =2.5e3
print(type(x))
print(type(y))
print(type(c))

#string
text = "hello world"
print(type(text))

#accessing character
print(text[1])
print(text[3])

#list
l1 = [1,2.5, "monday", "apple"]
print(l1)
print(type(l1))
print(l1[3])
l1.append("Litchi")
print(l1)

#Tuple
T = (1.2, "lovishhh", "brilliantide")
print(T)
print(type(T))
print(T[2])

#set
set = {1,3,4,4,5,2}
print(set)  #remove duplicates also
for x in set:
  print(x)

set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)

#dictionaries
dict = {
    "x" : "apple",
    "y" : "cat",
    "z" : 2
}
print(dict)
print(dict["y"])
print(dict.keys())
print(dict.values())
print(dict.items())

dict["z"] = 3
print(dict["z"])
print(dict)

num = "100"
num_int = int(num)
print(num_int) # Output: 100
print(type(num_int))
```

```python
#Indexing

a = "hello world"
print(a[2])
print(a[-1])

# indexing in list
list = [10,20,30,40,50]
print(list[3])
print(list[-2])

# indexing in list
list1 = [12,20, [10,40,60],9]
print(list1[-2])
print(list1[0])

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(nested_list[1][2]) # Output: 6
```

```python
#Slicing - Slicing allows extracting portions of sequences such as strings, lists, and tuples using the start:stop:step syntax.

# Basic Slicing Operations
# ● Start – The index from where slicing begins (inclusive).
# ● Stop – The index where slicing stops (exclusive).
# ● Step – The increment between indices (default is 1).

text = "hellolovely"
print(text[2:])
print(text[1:5])
print(text[:4])
print(text[5:])

# Slicing with Step Parameter
num = [1,2,3,4,5,6,6,7.7,8,9]
print(num[::2])
print(num[1::1])
print(num[::-1])  #reversing a list
num[1] = "apple"
print(num[1])  # replacing + modification of list
print(num)

# Using Slicing for String Manipulation
str = "lovely"
print(str[::-1])
print(str[::-2])

# Indexing and Slicing in Tuples
numbers = (10, 20, 30, 40, 50)
print(numbers[2]) # Output: 30
print(numbers[-1]) # Output: 50
print(numbers[:3])

nested_list = [[10, 20, 30], [40, 50, 60], [70, 80, 90]]
print(nested_list[1][2]) # Output: 60
print(nested_list[2][:2]) # Output: [70, 80]

nested_tuple = ((1, 2, 3), (4, 5, 6), (7, 8, 9))
print(nested_tuple[1][1]) # Output: 5
print(nested_tuple[0][:2]) # Output: (1, 2)
```

```python
# Arithmetic operators
a = 15
b = 2
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)

# comparison operators
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)

#logical operators
x = 15
print(x>10 and x<20)
print(x>13 or x<20)
print(not(x>10 and x<20))

#assigment operators
x = 10
x += 5
print(x) # Output: 15
x *= 2
print(x) # Output: 30
x -= 5
print(x) # Output: 25

# Membership Operators - in , not in
fruits = ["apple" , "Mango" , "litchi", "grapes"]
print("apple" in fruits)
print("Mango" not in fruits)
```
