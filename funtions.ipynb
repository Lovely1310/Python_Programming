```python
# List to Set
list1 = [1,2,3,4,5]
set1 = set(list1)
print(set1)

# String to Set
list2 = "monika"
set2 = set(list2)
print(set2)

# Tuple to Set
list3 = (1,2,3,4,5)
set3 = set(list3)
print(set3)
```

```python
list = [1,2,3,{3:4}]   # dict {3:40} throws an error because dict is mutable means (non hashable) but set only stores hashable value that's why it throws an error
f = set(list)          # set ek khud non hashable h or lekin vo store krta h hashable (immutable ) values ..jo unique ho
print(f)

list = [1,2,3,(3,4)]    # we use tuple in this case tuple is immutable (can't modified ) means (hashable) ...hashable stores unique value thats why it does not throws an error
e = set(list)
print(e)
```

```python
# let's consider the scenarion where I want to print
# "Hi, My name is <name>, I have enrolled in the <program>. Thank you!",
#Where Program is a variable which will change
# Including variable with the text

#method 1 : Using Comma separated values # Most Common

name = "monika"
program = "data science"
print('hi my name is',name,',i have enrolled in the',program,".thank you")

# method - 2,  fstring used
x = "systematic"
y = 12
print(f"the word is {x} & the no. is {y}")

#method - 3 use format()
x = "lovely"
y = "microsoft"
z = "very much"
print("hello I'm {} i am currently work in {} company, thank you {}".format(x , y,z))

#method - 4 use concatenation +
x = "lovely"
y = "microsoft"
z = "very much"
print("hello I'm " + x + " i am currently work in " + y + " company, thank you " + z)
```
